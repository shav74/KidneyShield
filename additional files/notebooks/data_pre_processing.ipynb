```python
import pandas as pd
import numpy as np
```

```python
#filtering only survey related data
filter_columns = [i for i in range(6,54)] + [79]
```

Importing csv files and parsing true, false and null values

```python
killinochchi_df = pd.read_csv('Killinochchi.csv',encoding='unicode_escape', true_values=['TRUE'], false_values=['FALSE'], na_values=['Null'], usecols=filter_columns)
colombo_df = pd.read_csv('Colombo.csv',encoding='unicode_escape', true_values=['TRUE'], false_values=['FALSE'], na_values=['Null'], usecols=filter_columns)
puttalam_df = pd.read_csv('Puttalam.csv',encoding='unicode_escape', true_values=['TRUE'], false_values=['FALSE'], na_values=['Null'], usecols=filter_columns)
kaligamauwa_df = pd.read_csv('Kaligamuawa.csv',encoding='unicode_escape', true_values=['TRUE'], false_values=['FALSE'], na_values=['Null'], usecols=filter_columns)
sumethradevi_df = pd.read_csv('Sumethradevi.csv',encoding='unicode_escape', true_values=['TRUE'], false_values=['FALSE'], na_values=['Null'], usecols=filter_columns)
```

```python
# merge all the dataFrames together
frames = [killinochchi_df, colombo_df, puttalam_df, kaligamauwa_df, sumethradevi_df]
ckd_df = pd.concat(frames)
```
